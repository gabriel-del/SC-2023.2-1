## Machine Learning

Machine Learning (Aprendizagem de Máquina) é um ramo da inteligência artificial (IA) e da ciência da computação que se concentra no uso de dados e algoritmos para imitar a maneira como os humanos aprendem, aprimorando gradualmente sua precisão.

Pode-se dividir os tipos de aprendizado em:

- `Supervisionado:` parte dos dados já está categorizada e é utilizada para treinamento

- `Não Supervisionado:` não se tem categorização nos dados usados

O problema tratado no projeto se enquadra no tipo de aprendizado supervisionado

Dentro dos problemas que envolvem aprendizado supervisionado, tem-se aqueles de:

- `Classificação:` os rótulos do conjunto de dados são categorias ou classes

Exs.: queremos classificar emails como limpos ou SPAM; se uma transação no cartão de crédito é fraudulenta ou não; se um investidor é conservador, moderado ou agressivo...

- `Regressão:` os rótulos representam valores/quantidades

Exs.: queremos prever quanto vale um imóvel; prever valor de uma ação; qual vai ser a demanda de determinado produto em um ponto de venda...

## KNN - K Nearest Neighbours

O KNN é um modelo usado para problemas de aprendizado supervisionado e pode ser utilizado para Classificação ou Regressão

Como o nome sugere, o KNN irá prever a categoria ou o valor de determinado objeto de estudo utilizando características dos "vizinhos mais próximos"

*Passo a passo:*

`1` - Dado um ponto de dados, calcula a distância de todos outros pontos no dataset;

    Esse cálculo pode ser feito por diversas métricas, sendo a mais comum a da distância euclidiana



`2` - Avalia os K pontos mais próximos (ex.: k = 3, 3 pontos mais próximos);

`3` - Se for de Classificação, calcula a moda entre os valores dos rótulos dos pontos mais próximos, se for de Regressão, calcula a média desses pontos.

### Exemplo hipotético 




1. Calcula a distância



2. Verifica os K mais próximos (k = 3)



3. *Regressão:* tira a média desses valores, *Classificação:* verifica o rótulo com maior ocorrência

```python
# Função para calcular distância euclidiana entre 2 pontos (w, x, y, z)
def dist_euclidiana(carteira_sem_classe, carteira_classificada):
    dist = ((carteira_classificada[0] - carteira_sem_classe[0])**2 + (carteira_classificada[1] - carteira_sem_classe[1])**2 + (carteira_classificada[2] - carteira_sem_classe[2])**2 + (carteira_classificada[3] - carteira_sem_classe[3])**2)**0.5
    return dist

# Função para calcular KNN de cada cliente de perfil desconhecido
def KNN(cart_no_class, dados_ref, k):

    # Criando lista com tuplas com Perfil, Distância Euclidiana
    distancias = [(perfil, dist_euclidiana(cart_no_class, cart)) for _, perfil, cart in dados_ref]

    # Ordenando lista utilizando a distância como critério
    distancias.sort(key= lambda x: x[1])

    # Definindo os k pontos mais próximos
    vizinhos = distancias[:k]

    # Listando classe dos pontos vizinhos
    classes_pontos_vizinhos = [perfil for perfil, _ in vizinhos]

    # Definindo classe mais frequente
    mais_frequente = max(set(classes_pontos_vizinhos), key=classes_pontos_vizinhos.count)
    
    return mais_frequente
```

```python
# Criando uma cópia da lista de investidores sem perfil e inserindo o perfil previsto

perfis_previstos = [KNN(pessoa[2], data, 3) for pessoa in no_class]

data_previsto = no_class.copy() 

for i in range(len(perfis_previstos)):
    data_previsto[i][1] = perfis_previstos[i]

data_previsto
```
